```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p(x,y,L):
    return np.sin(2*np.pi*(x/L-0.3))*np.sin(2*np.pi*(y/L-0.3)) + 1/4* np.sin(8*np.pi*(x/L-0.3))*np.sin(8*np.pi*(y/L-0.3))
```

```python
X,Y = np.meshgrid(np.linspace(0,32,100),np.linspace(0,32,100))

pL = p(X,Y,32)
```

```python
fig,ax = plt.subplots()
c = ax.contourf(X,Y,pL,levels=32)
ax.set_aspect("equal")
plt.colorbar(c)
```
